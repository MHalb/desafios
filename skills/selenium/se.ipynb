```python
from selenium import webdriver
from IPython.display import HTML
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
```

```python
options = webdriver.ChromeOptions()
options.add_argument("--no-sandbox")
engine = webdriver.Chrome(options=options)
```

```python
cobaia_url = "https://www.scrapethissite.com/pages/simple/"
engine.get(cobaia_url)
```

```python
engine.title
engine.current_url
engine.name
engine.page_source
#HTML(engine.page_source)
```

```python
element = engine.find_element(By.ID, "nav-faq").find_element(By.CLASS_NAME, "nav-link")
element_text =element.text
element.click()
engine.back()

print(f"{element_text=}")
```

```python
## get crountry name and info

# Isso aqui é muito mais lerdo que o necessário, talvez utilizando o bs4 em união com ele seria mais rápido que o proprio buscador de elementos do selenium

element = engine.find_elements(By.CLASS_NAME, "col-md-4.country")

items_ct = []

for item in element:
   items_ct.append({
        "name": item.find_element(By.CLASS_NAME, "country-name").text,
        "country_info": item.find_element(By.CLASS_NAME, "country-info").text
      }
   )
```

```python
html = BeautifulSoup(engine.page_source, "html.parser")

items_ct = []

for content in html.find_all("div", class_="col-md-4 country"):
   items_ct.append({
        "name": content.find("h3", "country-name").text.strip(),
        "country_info": content.find("div", "country-info").text.strip()
      }
   )

print(items_ct)
```
